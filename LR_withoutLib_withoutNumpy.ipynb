```python
X=list()
y=list()
for i in ds:
    X.append(i[0])
    y.append(i[1])
print(X)
print(y)

plt.plot(X,y,'go')
```

```python
def calc_fn(X,m,c):
    fn=list()
    for i in X:
        fn.append(i*m+c)
    return fn

def subtract(l1,l2):
    sub=list()
    for i,j in zip(l1,l2):
        sub.append(i-j)
    return sub

def squar(l1):
    sqr=list()
    for i in l1:
        sqr.append(i**2)
    return sqr
        
def multi(l1,l2):
    mult=list()
    for i,j in zip(l1,l2):
        mult.append(i*j)
    return mult

def summ(l1):
    sum=0
    for i in l1:
        sum=sum+i
    return sum

def cost_func(X,y,m,c):
    return (1/2*len(X))*summ(squar(subtract(calc_fn(X,m,c),y)))
```

```python
def gradient_desc(X,y,iterations):
    
    meansq=0
    c,m=0,0

    size=len(X)
    alpha=0.07
    for i in range(iterations):
        t1=c-alpha*((1/size)*summ(subtract(calc_fn(X,m,c),y)))
        t2=m-alpha*((1/size)*summ(multi((subtract(calc_fn(X,m,c),y)),X)))
        c=t1
        m=t2
    print(m,c)
    print(cost_func(X,y,m,c))
    
    
    YY=list()
    for i in X:
        YY.append(i*m+c)
    plt.plot(X,YY,'r-')
    
    
    return cost_func(X,y,m,c)
```
